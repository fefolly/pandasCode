```python
#import numpy as np
import pandas as pd

logInfoDf = pd.read_csv("pandas_tutorial_read.csv", delimiter=';', names=['my_datetime','event','country','user_id','source','topic'])

for i in logInfoDf.itertuples():
    j = 1  #i started with 1 because 0 would retrieve the index value which I do not need in this case
    while j <= 6:
        print(i[j], " ")
        j += 1
    source = i[5]
    print("the source of this one is:", source)
    print()
```

```python
for i in logInfoDf.itertuples():
    print("The time is: ", i[1])
    print("The name of the event is: ", i[2])
    print("The country it held in is: ", i[3])
    print("The user's is: ", i[4])
    print("The source is: ", i[5])
    print("The topic is: ", i[6])
    # In our case we need the the data in a variable which we can passe in parameter to the client
    #
    print()
```

```python
for i in logInfoDf.itertuples():
    time = i[1]                    #time is a simple variable that can be passed as a parameter to the client
    print("The time is: ", time)
    event = i[2]                    #event is a simple variable that can be passed as a parameter to the client
    print("The name of the event is: ", event)
    country = i[3]                   #country is a simple variable that can be passed as a parameter to the client
    print("The country it held in is: ", country)
    user = i[4]                       #user is a simple variable that can be passed as a parameter to the client
    print("The user's is: ", user)
    source = i[5]                     #source is a simple variable that can be passed as a parameter to the client
    print("The source is: ", source)
    topic = i[6]                       #topic is a simple variable that can be passed as a parameter to the client
    print("The topic is: ", topic)
    print()
```
